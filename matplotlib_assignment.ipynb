Matplotlib is a popular data visualization library in Python. It provides a wide range of tools for creating various types of plots, charts, and visualizations.

1. Line Plot
2. Bar Plot
3. Histogram
4. Scatter Plot
5. Pie Chart


A scatter plot is a type of plot that displays the relationship between two continuous variables. It represents individual data points as dots on a two-dimensional plane, where each dot corresponds to a specific value of the two variables being plotted.They are helpful in identifying patterns, clusters, or correlations between variables. It is created using the plt.scatter() function.  


```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.plot(x,y)
plt.title("Q2. Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to divide the plotting area into multiple smaller sections or subplots, each containing its own plot or visualization.

```python
import numpy as np

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x1,y1)

plt.subplot(2,2,2)
plt.plot(x2,y2)

plt.subplot(2,2,3)
plt.plot(x3,y3)

plt.subplot(2,2,4)
plt.plot(x4,y4)

plt.show()
```

A bar plot, also known as a bar chart, is a type of chart that uses rectangular bars to represent data. It is commonly used to compare and display categorical data or to show the relationship between a categorical variable and a numerical variable.

In a bar plot, the length or height of each bar corresponds to the value or frequency of the data it represents. The bars are typically arranged along the x-axis, with the y-axis representing the values being compared or measured.Bar plots are used for various purposes, including Comparison of Categories, Displaying Categorical Data, Trend Visualization, Highlighting Differences, Data Exploration and Presentation.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.xlabel("Company Name")
plt.ylabel("Profit")
plt.title("Companies and thir Profit")
plt.show()
```

A box plot, also known as a box-and-whisker plot, is a type of statistical chart used to display the distribution of a dataset. It provides a visual summary of key statistical measures and highlights the presence of outliers.

A box plot consists of several components:

1. Box: The box in the plot represents the interquartile range (IQR), which is the range between the 25th percentile (lower quartile or Q1) and the 75th percentile (upper quartile or Q3) of the data. The length of the box represents the spread of the middle 50% of the dataset.
2. Median Line: Inside the box, a line represents the median.
3. Whiskers: The whiskers extend from the box and represent the variability of the data outside the IQR. They typically extend to the most extreme data points that are still within a certain range from the box, often 1.5 times the IQR. Data points beyond this range are considered outliers and are displayed separately.
4. Outliers: Individual data points that fall outside the whiskers are considered outliers and are plotted as individual points.

Box plots are used for several purposes:

Visualizing Distribution

Identifying Skewness and Outliers

Comparing Distributions

Detecting Patterns or Relationships

Summarizing Data: Box plots provide a concise summary of the dataset, including the range, quartiles, and outliers. They are often used as part of statistical analysis or reporting to present a compact summary of key statistics.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.subplot(1,2,1)
plt.boxplot(box1)

plt.subplot(1,2,2)
plt.boxplot(box2)
plt.show()
```
